```python
alpha = float(0.0001)
epsilon = float(0.1)
gamma = float(0.8)
numTraining = int(20)
w = [0.5,0.2, 0.1]
```

```python
#this function get the next row and colume based on the policy with probability 0.8
#and perpendicular on the policy with probability 0.2

#if the function will take the perpendicular direction
#it will choose the either of the two perpendicular direction on 50% probability
def get_next_r_c(r, c):
    prob = random()
    #80% of the time will take the policy action
    if prob <= 0.8: 
        if policy[r][c] == 'up':
            rnext = r - 1
            cnext = c
        elif policy[r][c] == 'left':
            rnext = r
            cnext = c - 1
        elif policy[r][c] == 'right':
            rnext = r
            cnext = c + 1
        elif policy[r][c] == 'down':
            rnext = r + 1
            cnext = c
            
    #20% of the time will make a perpendiclar action on the policy action
    else:
        if policy[r][c] == 'up' or policy[r][c] == 'down':
            prob2 = random()
            if prob2>0.5:
                #left
                rnext = r
                cnext = c - 1
            else:
                #right
                rnext = r
                cnext = c + 1
                
        elif policy[r][c] == 'right' or policy[r][c] == 'left':
            prob2 = random()
            if prob2>0.5:
                #up
                rnext = r - 1
                cnext = c
            else:
                #down
                rnext = r + 1
                cnext = c
                
    return rnext, cnext
```

```python
def recursive_approximation(r, c):
    #updates delta values recursively for each step
    if policy[r][c] == 'exit':
        if positive_reward_position_row == r and positive_reward_position_col == c:
            return 1
        else:
            return -1
    
    rnext, cnext = get_next_r_c(r, c)
    
    while rnext >= rows or rnext < 0 or cnext >= cols or cnext < 0 or policy[rnext][cnext] == 'obstacle':
        rnext, cnext = get_next_r_c(r, c)
    
    value = recursive_approximation(rnext,cnext) + one_step_reward
    delta_rule(r,c, value)
    return value
```

```python
def approximate_value(r,c):
    return w[0]+w[1]*c+w[2]*r
```

```python
def delta_rule(r,c, current_observed_value):
        w[0] = w[0] + alpha * (current_observed_value - approximate_value(r,c))
        w[1] = w[1] + alpha * (current_observed_value - approximate_value(r,c)) * c
        w[2] = w[2] + alpha * (current_observed_value - approximate_value(r,c)) * r
```

```python
def function_approximation():
    for i in range(iterations):
        r = randint(0, rows - 1)
        c = randint(0, cols - 1)
        while policy[r][c] == 'obstacle' or policy[r][c] == 'exit':
            r = randint(0, rows - 1)
            c = randint(0, cols - 1)
        recursive_approximation(r, c)

```

```python
def view_utility_map():
    utility_map = np.zeros((rows,cols))
    for r in range(rows):
        for c in range(cols):
            utility_map[r,c] = w[0]+w[1]*c+w[2]*r
    print(utility_map)
```
