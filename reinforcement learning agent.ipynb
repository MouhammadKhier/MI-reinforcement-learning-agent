```python
from random import randint
from random import random
from random import seed
import numpy as np
```

```python
def get_next_r_c(r, c):
    prob = random()
    #80% of the time will take the policy action
    if prob <= 0.8: 
        if policy[r][c] == 'up':
            rnext = r - 1
            cnext = c
        elif policy[r][c] == 'left':
            rnext = r
            cnext = c - 1
        elif policy[r][c] == 'right':
            rnext = r
            cnext = c + 1
        elif policy[r][c] == 'down':
            rnext = r + 1
            cnext = c
            
    #20% of the time will make a perpendiclar action on the policy action
    else:
        if policy[r][c] == 'up' or policy[r][c] == 'down':
            prob2 = random()
            if prob2>0.5:
                #left
                rnext = r
                cnext = c - 1
            else:
                #right
                rnext = r
                cnext = c + 1
                
        elif policy[r][c] == 'right' or policy[r][c] == 'left':
            prob2 = random()
            if prob2>0.5:
                #up
                rnext = r - 1
                cnext = c
            else:
                #down
                rnext = r + 1
                cnext = c
                
    return rnext, cnext
```

```python
def recursive_learning(r, c):
    if policy[r][c] == 'exit':
        return Vs[r][c]
    
    rnext, cnext = get_next_r_c(r, c)
    
    while rnext >= rows or rnext < 0 or cnext >= cols or cnext < 0 or policy[rnext][cnext] == 'obstacle':
        rnext, cnext = get_next_r_c(r, c)
    
    value = recursive_learning(rnext,cnext) + one_step_reward
    
    Vs[r][c] = Vs[r][c] + value
    count[r][c] += 1
    
    return value
```

```python
def TRlearn():
    r = randint(0, rows - 1)
    c = randint(0, cols - 1)
    while policy[r][c] == 'obstacle' or policy[r][c] == 'exit':
        r = randint(0, rows - 1)
        c = randint(0, cols - 1)
    
    recursive_learning(r, c)
```

```python
def main():
    for i in range(100000):
        #print("pass number", i+1)
        TRlearn()
        
    for r in range(rows):
        for c in range(cols):
            if policy[r][c] != 'exit' and count[r][c] != 0:
                Vs[r][c] = Vs[r][c]/count[r][c]
                
    values = np.array(Vs)
    print(values)
```
